```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Create folder for samples
SAMPLES_DIR = r'C:\Users\srika\gw_wd\results\figures'
os.makedirs(SAMPLES_DIR, exist_ok=True)

# Load your test data
X_test = np.load(r'C:\Users\srika\gw_wd\data\dataset_qtransform\X_test.npy')
y_test = np.load(r'C:\Users\srika\gw_wd\data\dataset_qtransform\y_test.npy')

print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
print(f"Signal samples in test: {np.sum(y_test == 1)}")
print(f"Noise samples in test: {np.sum(y_test == 0)}")

# Find signal and noise examples
signal_indices = np.where(y_test == 1)[0]
noise_indices = np.where(y_test == 0)[0]

print(f"\nSignal indices: {signal_indices[:5]}")
print(f"Noise indices: {noise_indices[:5]}")

# Save 2 best signal examples
for i in range(min(2, len(signal_indices))):
    idx = signal_indices[i]
    img = X_test[idx]
    
    # Save as PNG
    plt.imsave(
        os.path.join(SAMPLES_DIR, f'sample_signal_{i+1}.png'),
        img
    )
    print(f"✅ Saved sample_signal_{i+1}.png")

# Save 2 best noise examples
for i in range(min(2, len(noise_indices))):
    idx = noise_indices[i]
    img = X_test[idx]
    
    # Save as PNG
    plt.imsave(
        os.path.join(SAMPLES_DIR, f'sample_noise_{i+1}.png'),
        img
    )
    print(f"✅ Saved sample_noise_{i+1}.png")

print(f"\n✅ All sample images saved to: {SAMPLES_DIR}")

# Verify files exist
import glob
samples = glob.glob(os.path.join(SAMPLES_DIR, 'sample_*.png'))
print(f"\nFiles created ({len(samples)}):")
for s in sorted(samples):
    size = os.path.getsize(s) / 1024
    print(f"  • {os.path.basename(s)} ({size:.1f} KB)")
```
